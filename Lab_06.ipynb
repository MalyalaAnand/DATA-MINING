```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data={'category':['A','B','C','A','B']}
df=pd.DataFrame(data)
```

```python
df
```

```python
Label_Encoder=LabelEncoder()
df['label_encoded']=Label_Encoder.fit_transform(df['category'])
df
```

```python
One_Hot_Encoder=pd.get_dummies(df['category'],prefix='category')
df=pd.concat([df,One_Hot_Encoder],axis=1)
df
```

```python
from datetime import datetime
data={'Name':['Alice','Bob','Charlie'],'Date_of_birth':['1990-05-15','1985-09-20','1995-02-10']}
df=pd.DataFrame(data)
df
```

```python
df['Date_of_birth']=pd.to_datetime(df['Date_of_birth'])
df
```

```python
current_year=datetime.now().year
df['Age']=current_year-df['Date_of_birth'].dt.year
df
```

```python
data={'Category':['A','A','B','B','C'],'value':[10,20,10,30,40]}
df=pd.DataFrame(data)
df
```

```python
grouped_df=df.groupby('Category').agg({'value':['sum','mean']})
grouped_df
```

```python
pivot=df.pivot_table(index='Category',values='value',aggfunc=['sum','mean'])
pivot
```
