```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Exam_Scores': [35, 50, 55, 65, 70, 75, 80, 90, 95],
    'Sleep_Hours': [8, 7, 7, 6, 6, 5, 5, 4, 4]
}
```

```python
print(type(data))
```

```python
data.keys()
```

```python
data.items()
```

```python
df=pd.DataFrame(data)
```

#Calculate Correlation Matrix

```python
df['Hours_Studied'].corr(df['Exam_Scores'])
```

```python
correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Hours_Studied',y='Exam_Scores',data=df)
plt.title('Hours_Studied vs Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Hours_Studied',y='Sleep_Hours',data=df)
plt.title('Hours_Studied vs Sleep Hours')
plt.xlabel('Hours Studied')
plt.ylabel('Sleep Hours')
plt.show()
```
